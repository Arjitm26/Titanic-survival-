```python
def agefill(col):
  age = col[0]
  pclass = col[1]
  if pd.isna(age):
    if pclass == 1:
      return 37
    elif pclass == 2:
      return 28
    else:
      return 24
  else:
    return age    
```

```python
df01['Age'] = df01[['Age','Pclass']].apply(agefill,axis = 1)
sn.heatmap(df01.isnull(),yticklabels= False,cbar = False)
```

```python
X = df01
```

```python
X.head()
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,x_test,Y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
predictions = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

accuracy = confusion_matrix(y_test,predictions)
```

```python
lr.score(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

print('Accuracy score: ',accuracy_score(y_test,predictions))
print('Precision is: ', precision_score(y_test,predictions,average = 'macro'))
print('Recall is: ', recall_score(y_test,predictions,average = 'weighted'))
print('F1 is: ', f1_score(y_test,predictions,average = 'macro'))
```

```python
sn.heatmap(accuracy,
            annot=True,
            fmt='g',
            )
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
```

```python
import tensorflow as tf

ann = tf.keras.models.Sequential()

ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
ann.fit(X_train,Y_train,batch_size=32,epochs = 100)
```
